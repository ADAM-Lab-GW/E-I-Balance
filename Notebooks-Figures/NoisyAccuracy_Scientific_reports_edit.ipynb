# Noise Accuracy Comparison (Fig 2b-e)
The noise data is loaded up more easily by aggregating the parameter files and accuracy files from each (no need for additional post processing like other metrics like the van rossum distance).

```python
from analysisUtil import *
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
noiseLevel = []
accAvg = []
colors = []
excHidRat = []
# colorDefs = {
#     50: '#99CCFF',  # light sky blue
#     70: '#66CCFF',  # pastel blue
#     75: '#33CCFF',  # light steel blue
#     80: '#3399CC',  # medium blue
#     90: '#9966CC',  # light purple
#     95: '#6633CC', # Purple
#     100: '#000000'   # Black
# }
colorDefs = {
    50: '#E6CCFF',  # light lavender
    70: '#CC99FF',  # lavender
    75: '#B366FF',  # rich lavender
    80: '#9933FF',  # dark lavender
    90: '#8000FF',  # deep lavender
    95: '#6600CC', # dark purple
    100: '#000000'   # Black
}
colorDefs = {
    50: '#BEBBEC',  # light lavender
    70: '#BA94D8',  # lavender
    75: '#BF72BB',  # rich lavender
    80: '#8E4198',  # dark lavender
    90: '#442F8E',  # deep lavender
    95: '#204B80', # dark purple
    100: '#000000'   # Black
}

for t in range(100,300):
    for e in ['050','070','075','080','090','095','100']:
        simTag = 'sparse'+e+'_noise_final_'+str(t)+'_t'+str((t-100)%8)
        try:
            param = getWeightFromUniqueId(simTag+'_surrParams', objFolder='/media/adamlab/My Passport/LIF Analysis Data/FashionMNIST/raw_data/sparse_noise_final_surrParams')
            acc = getWeightFromUniqueId(simTag+'_accList', objFolder='/media/adamlab/My Passport/LIF Analysis Data/FashionMNIST/raw_data/sparse_noise_final_acc')

            accAvg.append(sum(acc[-10:])/10)
            noiseLevel.append(param.w1NoiseSTD)
            excHidRat.append(int(e))
            colors.append(colorDefs[int(e)])
        except:
            print('skipped')

        try:
            param = getWeightFromUniqueId(simTag+'_surrParams', objFolder='/media/adamlab/My Passport/LIF Analysis Data/FashionMNIST/raw_data/sparse_noise_final_surrParams')
            acc = getWeightFromUniqueId(simTag+'_accList', objFolder='/media/adamlab/My Passport/LIF Analysis Data/FashionMNIST/raw_data/sparse_noise_final_acc')

            accAvg.append(sum(acc[-10:])/10)
            noiseLevel.append(param.w1NoiseSTD)
            excHidRat.append(int(e))
            colors.append(colorDefs[int(e)])
        except:
            print('skipped')
        
```

```python
excHidRat
```

```python
ratios = [50,70,75,80,90,95,100]
noiseLevels = [0.00001, 0.0001, 0.00015, 0.0002, 0.0003, 0.0004, 0.0005, 0.0006, 0.0007, 0.0008, 0.0009, 0.001]
allTrial = {}
for r in ratios:
    allTrial[r] = {}
    for n in noiseLevels:
        allTrial[r][n] = []
        for i in range(len(accAvg)):
            if excHidRat[i]!=r: continue
            if noiseLevel[i]!=n: continue
            allTrial[r][n].append(accAvg[i])
    
```

```python
fig, axs = plt.subplots(1,4, figsize=(12,3))
colorDefs = {
    50: '#BEBBEC',  # light lavender
    70: '#BA94D8',  # lavender
    75: '#BF72BB',  # rich lavender
    80: '#8E4198',  # dark lavender
    90: '#442F8E',  # deep lavender
    95: '#204B80', # dark purple
    100: '#000000'   # Black
}
colorDefs = {
100:'#000000',
 95:'#8D13A3',
 90:'#7C2ED1',
 80:'#6E51EA',
 75:'#727CE8',
 70:'#91B1E8',
 50:'#AED5EB'
}

ratios.reverse()
for r in ratios:
    xScatter = []
    yScatter = []
    xLine = []
    yLine = []
    for n in noiseLevels:
        xScatter.extend([n]*len(allTrial[r][n]))
        yScatter.extend([t*100 for t in allTrial[r][n]])
        xLine.append(n)
        yLine.append(100*sum(allTrial[r][n])/len(allTrial[r][n]))
    axs[0].scatter(xScatter,yScatter, color=colorDefs[r], alpha=0.2)
    axs[0].plot(xLine,yLine, color=colorDefs[r], label='{:d}:{:d}'.format(r,100-r))
axs[0].legend()
axs[0].set_xticks([0.0000,0.0002,0.0004,0.0006,0.0008,0.001])
axs[0].set_xticklabels(['0.0000','0.0002','0.0004','0.0006','0.0008','0.001'],rotation=30)
axs[0].set_xlabel('Noise STD')
axs[0].set_ylabel('Average Accuracy\n(Last 10 Epochs)')
axs[0].set_ylim(0,100)

noisePlots = [None, 0.0002, 0.0004, 0.0006]
ratioPlots = [75,80,90,95,100]
for i in range(1,4):
    for j in range(len(ratioPlots)):
        r = ratioPlots[j]
        n = noisePlots[i]
        diff = 100*(np.average(allTrial[r][n])-np.average(allTrial[100][n]))
        std = np.std([t*100 for t in allTrial[r][n]])
        axs[i].bar(j, diff, color=colorDefs[r])
        (_,caps,_) = axs[i].errorbar(j, diff, yerr=std, color='black', capsize=3)
        for cap in caps:
            cap.set_markeredgewidth(1)
            
        t_statistic, p_value = stats.ttest_ind(allTrial[r][n], allTrial[100][n])
        if p_value<0.001:
            axs[i].text(j, 5, '✱✱✱', horizontalalignment='center', color=colorDefs[r])
        elif p_value<0.01:
            axs[i].text(j, 5, '✱✱', horizontalalignment='center', color=colorDefs[r])
        elif p_value<0.05:
            axs[i].text(j, 5, '✱', horizontalalignment='center', color=colorDefs[r])

    axs[i].plot([-1,5],[0,0],color='black')
    axs[i].set_xticks(range(len(ratioPlots)))
    axs[i].set_xticklabels(['{:d}:{:d}'.format(x, 100-x) for x in ratioPlots], rotation=30)
    axs[i].set_xlim(-0.5, 4.5)
    axs[i].set_xlabel('E:I Ratio')
    axs[i].set_ylabel('Difference with 100:0 Network')
    
axs[1].set_ylim(-21,8)
#axs[1].text(4,-5,'Noise STD=0.0002',horizontalalignment='right')
axs[2].set_ylim(-21,8)
#axs[2].text(4,-5,'Noise STD=0.0002',horizontalalignment='right')
axs[3].set_ylim(-21,8)
#axs[3].text(4,-19,'Noise STD=0.0002',horizontalalignment='right')
            
for i in range(4):
    axs[i].grid()
    axs[i].set_axisbelow(True)
plt.tight_layout()

plt.savefig('./noise_plots/combined_accV15_SR.svg')
```
